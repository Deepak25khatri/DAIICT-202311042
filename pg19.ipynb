```python
#Example - Plotting x and y points - Line

import matplotlib.pyplot as plt
import numpy as np
xpoint = np.array([0,4])
ypoint = np.array([0,50])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
#Example - Plotting x and y points - Without Line

xpoint = np.array([0,4])
ypoint = np.array([0,50])
plt.plot(xpoint,ypoint, 'o')
plt.show()
```

```python
#Example - Plotting Multiple Points

xpoint = np.array([1,2,6,8])
ypoint = np.array([3,9,1,10])
plt.plot(xpoint,ypoint)
plt.show()
```

```python
#Example - Markers - Format Strings fmt

ypoint = np.array([3,9,1,10])
plt.plot(ypoint,"o:r")
plt.show()
```

```python
#Example - Markers Size

ypoint = np.array([3,9,1,10])
plt.plot(ypoint,marker="o",ms=20)
plt.show()
```

```python
#Example - Markers Color

ypoint = np.array([3,9,1,10])
plt.plot(ypoint,marker="o",ms=20,mec = 'g')
plt.show()
```

```python
#Example - Markers  face Color

ypoint = np.array([3,9,1,10])
plt.plot(ypoint,marker="o",ms=20,mec = 'g',mfc = 'r')
plt.show()
```

```python
#Example - Linestyle

ypoint = np.array([3,9,1,10])
plt.plot(ypoint,marker="o",ms=20,mec = 'g',mfc = 'r', linestyle="dotted")
plt.show()
```

```python
#Example - Line Width

ypoint = np.array([3,9,1,10])
plt.plot(ypoint,marker="o",ms=20,mec = 'g',mfc = 'r', linestyle="dotted",linewidth="20.5")
plt.show()
```

```python
#Example - Multiple Line 

xpoint = np.array([1,4,8,5])
ypoint = np.array([2,1,8,9])
plt.plot(xpoint)
plt.plot(ypoint)
plt.show()
```

```python
#Example - Label

x = np.array([80,85,90,95,100,105,110,115,120])
y = np.array([240,250,260,270,280,290,300,310,320])
plt.plot(x,y)
plt.xlabel("Average pulse")
plt.ylabel("calorie burn")

plt.show()
```

```python
#Example - Title

x = np.array([80,85,90,95,100,105,110,115,120])
y = np.array([240,250,260,270,280,290,300,310,320])
plt.plot(x,y)
plt.title("sports watch data")
plt.xlabel("Average pulse")
plt.ylabel("calorie burn")

plt.show()

```
